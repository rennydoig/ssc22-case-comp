```python
## Import forecasted data
arima_d_df = pd.read_csv('arima_d_forecasted.csv')
arima_u_df = pd.read_csv('arima_u_forecasted.csv')
lmm_d_df = pd.read_csv('lmm_d_forecasted.csv')
lmm_u_df = pd.read_csv('lmm_u_forecasted.csv')
```

```python
## After cross-validattion, we found that arima:lmm = [0.9, 0.1] is the best set of weights

ensemble_d_df = arima_d_df.copy()
ensemble_d_df.iloc[:, 5:] = arima_d_df.iloc[:, 5:]*0.9 + lmm_d_df.iloc[:, 5:]*0.1
ensemble_d_df.head()
```

```python
ensemble_u_df = arima_u_df.copy()
ensemble_u_df.iloc[:, 5:] = arima_u_df.iloc[:, 5:]*0.9 + lmm_u_df.iloc[:, 5:]*0.1
ensemble_u_df.head()
```

```python
ensemble_d_df.to_csv("ensemble_d_forecasted.csv")
ensemble_u_df.to_csv("ensemble_u_forecasted.csv")
```
